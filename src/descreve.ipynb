# 2. janela_hamming
ela vai suavizar as bordas de cada trecho do áudio. sem ela teriamos cortes bruscos no sinal geram vazamento espectral aparecem frequências falsas
Por que esses valores (0.54 e 0.46)?
- Escolhidos empiricamente
- Reduzem vazamento espectral
- Mantêm boa resolução de frequência

A janela de Hamming é usada para reduzir o vazamento espectral causado
por cortes abruptos no sinal. Ao aplicar a FFT em pequenos trechos (frames),
as bordas do sinal podem gerar descontinuidades, introduzindo frequências falsas.

 A função:
w[n] = 0.54 - 0.46 * cos(2πn / (N-1))

suaviza essas bordas, reduzindo esse efeito.

As constantes 0.54 e 0.46 foram empiricamente escolhidas para equilibrar:
- baixa distorção espectral
- boa resolução de frequência

Isso melhora a qualidade do espectrograma e, consequentemente,
o desempenho do modelo de classificação.
